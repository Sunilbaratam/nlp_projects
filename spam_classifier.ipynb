```python
import pandas as pd
import nltk
import re
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('SMSSpamCollection.txt', sep = '/t',
                  names = ["label", "message"])

```

```python
data.head()
```

```python
#Data cleaning and preprocessing
prep = WordNetLemmatizer()
stem = PorterStemmer()
corpus = []
for i in range(0, len(data)):
    output = re.sub('[^a-zA-Z]', ' ', str(data['message'][i]))
    output = output.lower()
    output = output.split()
    
    output = [prep.lemmatize(each) for each in output if each not in set(stopwords.words('english')) ]
    ##lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    output = ' '.join(output)
    corpus.append(output)
```

```python
corpus
```

```python
#creating bagofwords

from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer(max_features=2500)
x = v.fit_transform(corpus).toarray()

y = pd.get_dummies(data['label'])
#y = y.iloc[;,-1].values
y=y.iloc[:,1].values
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)


```

```python
# Training model using Naive bayes classifier

from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)
```
